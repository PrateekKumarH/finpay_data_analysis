```python
import pandas as pd
df = pd.read_csv("/Users/prateekprajapati/Documents/finPay_analytics_data/raw/transactions_large.csv")
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df['transaction_ts'] = pd.to_datetime(df['transaction_ts'])
```

```python
df['date'] = df['transaction_ts'].dt.date
df['year'] = df['transaction_ts'].dt.year
df['month'] = df['transaction_ts'].dt.month
df['month_name'] = df['transaction_ts'].dt.month_name()
df['weekday'] = df['transaction_ts'].dt.day_name()
df['hour'] = df['transaction_ts'].dt.hour
```

```python
df.head()
```

```python
df['is_promo_used'] = df['is_promo_used'].map({'Yes': 1, 'No': 0})
```

```python
df
```

```python
dim_country=df[['country']].drop_duplicates().reset_index(drop=True)
dim_country['country_id']=dim_country.index+1
dim_country
```

```python
dim_device=df[['device_type']].drop_duplicates().reset_index(drop=True)
dim_device['device_id']=dim_device.index+1
dim_device
```

```python
dim_payment=df[['payment_method']].drop_duplicates().reset_index(drop=True)
dim_payment['payment_method_id']=dim_payment.index+1
dim_payment
```

```python
dim_user=df[['user_id']].drop_duplicates().reset_index(drop=True)
dim_user
```

```python
dim_merchant=df[['merchant_id', 'category']].drop_duplicates()
dim_merchant
```

```python
dim_date=df[['date', 'year', 'month', 'month_name', 'weekday', 'hour']].drop_duplicates()
dim_date
```

```python
df.info()
```

```python
fact=df.merge(dim_country, on='country', how='left')
fact=fact.merge(dim_device, on='device_type', how='left')
fact=fact.merge(dim_payment, on='payment_method', how='left')
```

```python
dim_country
```

```python
df
```

```python
fact_transactions=fact[['transaction_id','user_id', 'merchant_id', 'country_id', 'device_id', 'payment_method_id','date', 'amount', 'status','is_promo_used']]
```

```python
fact_transactions
```

```python
dim_country.to_csv('dim_country.csv', index=False)
dim_device.to_csv('dim_device.csv', index=False)
dim_payment.to_csv('dim_payment.csv', index=False)
dim_user.to_csv('dim_user.csv', index=False)
dim_merchant.to_csv('dim_merchant.csv', index=False)
dim_date.to_csv('dim_date.csv', index=False)
fact_transactions.to_csv('fact_transactions.csv', index=False)
```
